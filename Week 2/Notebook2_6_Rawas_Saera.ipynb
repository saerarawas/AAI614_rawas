```python
# Create a new column, pre_tip_amount
df['pre_tip_amount'] = (
    df
    [['fare_amount', 'extra', 'mta_tax', 'tolls_amount',
      'improvement_surcharge', 'congestion_surcharge']]
    .sum(axis='columns')
)
```

```python
# Create a new column, tip_percentage, showing the percent of the fare_amount that people tipped
#df['tip_percentage'] = df['tip_amount'] / df['pre_tip_amount']
import pandas as pd
#Convert 'tip_amount' and 'pre_tip_amount' to numeric, coercing errors to NaN
df['tip_percentage'] = pd.to_numeric(df['tip_amount'], errors='coerce') / pd.to_numeric(df['pre_tip_amount'], errors='coerce')
```

```python
# What was the overall tip percentage?
df['tip_percentage'].mean()
```

```python
# How many times did people tip more than 100%?
(df['tip_percentage'] > 1).value_counts()
```

```python
# What percent of taxi riders give no tip at all?
(df['tip_percentage'] == 0).value_counts(normalize=True)
```

```python
# On which day of the week do people tip the greatest percentage?
# Mon = 0, Tues = 1, Wed = 2, Thu = 3, Fri = 4, Sat = 5, Sun = 6

#df.groupby(df['tpep_pickup_datetime'].dt.day_of_week)['tip_percentage'].mean().sort_values(ascending=False)
import pandas as pd

# Assuming df is your DataFrame
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'], errors='coerce')  # Convert to datetime, handling errors
# errors='coerce' will replace invalid dates with NaT (Not a Time)

# Now your original code should work
df.groupby(df['tpep_pickup_datetime'].dt.day_of_week)['tip_percentage'].mean().sort_values(ascending=False)
```

```python
# At which hour do people tip the greatest percentage?
df.groupby(df['tpep_pickup_datetime'].dt.hour)['tip_percentage'].mean().sort_values(ascending=False)
```

```python
# Do people tip more, on average, in January or July?
df.groupby(df['tpep_pickup_datetime'].dt.month)['tip_percentage'].mean().sort_values(ascending=False)
```

```python
# What was the 1-day period in our data set when people tipped the greatest percentage?
df = df.set_index('tpep_pickup_datetime')
df.resample('1D')['tip_percentage'].mean().sort_values(ascending=False).head(10)
```

```python
df = df.sort_index()

# Convert index to datetime if it's not already
df.index = pd.to_datetime(df.index)

# Use .loc with boolean indexing to select the desired date ranges
df = pd.concat([
    df.loc[(df.index >= '2019-01-01') & (df.index <= '2019-01-31')],
    df.loc[(df.index >= '2019-07-01') & (df.index <= '2019-07-31')]
])
```

```python
df.resample('1M')['tip_percentage'].mean().dropna()
```

```python
import pandas as pd

# Calculate the total number of rides
total_rides = len(df)

# Calculate the number of rides with 0 tip
no_tip_rides = len(df[df['tip_amount'] == 0])

# Calculate the percentage of riders who don't tip
no_tip_percentage = (no_tip_rides / total_rides) * 100

print(f"Percentage of riders who don't tip: {no_tip_percentage:.2f}%")
```

3. Looking only at dates in January and July, on what week did passengers tip the greatest percentage?

```python
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tip_percentage'] = (df['tip_amount'] / df['total_amount']) * 100

# Filter for January and July
jan_jul_df = df[(df['tpep_pickup_datetime'].dt.month == 1) | (df['tpep_pickup_datetime'].dt.month == 7)]

# Group by week and calculate average tip percentage
weekly_tips = jan_jul_df.groupby(jan_jul_df['tpep_pickup_datetime'].dt.isocalendar().week)['tip_percentage'].mean()

# Find the week with the highest average tip percentage
best_week = weekly_tips.idxmax()

print(f"The week with the greatest average tip percentage in January and July is: {best_week}")
```
